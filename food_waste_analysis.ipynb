## Descriptive Analysis

```python
import pandas as pd

# Load your dataset
df = pd.read_csv("food_waste.csv")

# Disable scientific notation for readability
pd.set_option('display.float_format', '{:.2f}'.format)

# 1. Basic structure
print("Shape of the dataset:", df.shape)
print("\nColumn names:", df.columns.tolist())

# 2. Summary statistics for all numeric columns
print("\n--- Summary Statistics ---")
print(df.describe())

# 3. Mean values for per capita waste
print("\n--- Average Per Capita Waste (kg/year) ---")
columns_of_interest = [
    'combined figures (kg/capita/year)',
    'Household estimate (kg/capita/year)',
    'Retail estimate (kg/capita/year)',
    'Food service estimate (kg/capita/year)'
]
print(df[columns_of_interest].mean())

# 4. Country with highest total household waste (tonnes)
highest_household = df.loc[df['Household estimate (tonnes/year)'].idxmax()]
print("\n--- Country with Highest Household Waste (tonnes) ---")
print(highest_household[['Country', 'Household estimate (tonnes/year)']])

# 5. Region-wise average per capita waste
print("\n--- Region-wise Average Per Capita Waste ---")
region_group = df.groupby('Region')[columns_of_interest].mean().sort_values(
    by='combined figures (kg/capita/year)', ascending=False)
print(region_group)

```
